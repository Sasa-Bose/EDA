'M' as 1000 KB, and
'k' as KB directly (i.e., just removing 'k')

If we simply replace '20M' with '20000', which assumes:

20M → 20 * 1000 = 20000 KB, which is what your code achieves via simple string manipulation.
Potential Pitfall:
If you ever have fractional MB values like '1.5M', your approach ('1.5M'.replace('M', '000')) turns it into '1.5000', which becomes 15000.0 when converted to float—not accurate (should be 1500.0 KB).
✅ Use 1024 if you're doing technical work, memory/storage calculations, or aligning with how computers measure data.

✅ Use 1000 if you're following marketing standards or working with datasets that define 1 MB as 1000 KB (often the case in app store data, for example).

```python
df_copy['Size'].isnull().sum()
```

```python
df_copy['Size'] = df_copy['Size'].replace('Varies with device', np.nan)
df_copy['Size'] = df_copy['Size'].astype(str).str.strip().str.replace(',', '.')
```

```python
def convert_to_kb(size):
    try:
        if 'M' in size:
            return float(size.replace('M', '')) * 1024
        elif 'k' in size:
            return float(size.replace('k', ''))
        else:
            return np.nan
    except:
        return np.nan

df_copy['Size'] = df_copy['Size'].apply(convert_to_kb)
```

1st I handled Intalls and Price columns using the above code later I realized, it has Potential Issue:
If df_copy['Installs'] or df_copy['Price'] has non-string values (like floats, NaN, or ints), this will raise an error like: 
AttributeError: Can only use .str accessor with string values!

```python
# Safer Version to handle Installs and Price
chars_to_remove=['+',',','$']
cols_to_clean=['Installs','Price']
for item in chars_to_remove:
    for col in cols_to_clean:
        df_copy[col] = df_copy[col].astype(str).str.replace(item, '')
```

# .astype(str)
Converts all the values in that column to strings, regardless of their original type (like float, int, NaN, etc.).

This is important because .str.replace() only works on strings. If the column has numbers or missing values, you'll get an error without this.

```python
df_copy['Installs'] = pd.to_numeric(df_copy['Installs'], errors='coerce')
df_copy['Price'] = pd.to_numeric(df_copy['Price'], errors='coerce')
```

'coerce': replaces invalid values with NaN.

# Which Category has largest number of installations??

```python
df_cat_installs = df_copy.groupby(['Category'])['Installs'].sum().sort_values(ascending = False).reset_index()
df_cat_installs.Installs = df_cat_installs.Installs/1000000000# converting into billions
df2 = df_cat_installs.head(10)
plt.figure(figsize = (14,10))
sns.set_context("talk")
sns.set_style("darkgrid")

ax = sns.barplot(x = 'Installs' , y = 'Category' , data = df2 )
ax.set_xlabel('No. of Installations in Billions')
ax.set_ylabel('')
ax.set_title("Most Popular Categories in Play Store", size = 20)
```

Insights
Out of all the categories "GAME" has the most number of Installations.
With almost 35 Billion Installations GAME is the most popular Category in Google App store

# What are the Top 5 most installed Apps in Each popular Categories ??

```python
dfa = df_copy.groupby(['Category' ,'App'])['Installs'].sum().reset_index()
dfa = dfa.sort_values('Installs', ascending = False)
apps = ['GAME', 'COMMUNICATION', 'PRODUCTIVITY', 'SOCIAL' ]
sns.set_context("poster")
sns.set_style("darkgrid")

plt.figure(figsize=(40,30))

for i,app in enumerate(apps):
    df2 = dfa[dfa.Category == app]
    df3 = df2.head(5)
    plt.subplot(4,2,i+1)
    sns.barplot(data= df3,x= 'Installs' ,y='App' )
    plt.xlabel('Installation in Millions')
    plt.ylabel('')
    plt.title(app,size = 20)
    
plt.tight_layout()
plt.subplots_adjust(hspace= .3)
plt.show()
```

Insights
Most popular game is Subway Surfers.
Most popular communication app is Hangouts.
Most popular productivity app is Google Drive.
Most popular social app is Instagram.

# How many apps are there on Google Play Store which get 5 ratings??

```python
rating = df_copy.groupby(['Category','Installs', 'App'])['Rating'].sum().sort_values(ascending = False).reset_index()

toprating_apps = rating[rating.Rating == 5.0]
print("Number of 5 rated apps",toprating_apps.shape[0])
toprating_apps.head(1)
```
